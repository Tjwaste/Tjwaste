```python
import random
lgbtq=['lesbian','gay','bisexual','transgender','queeer']
action=['rub','bounce','lick','smell','play','bite','suck']
connective=['on','with','into','in','the','out']
food=['sausage','banana','fish','cake','milkshake','cucumber','flower','ham']
f'{random.choice(lgbtq)} {random.choice(action)} {random.choice(connective)} {random.choice(food)}'
```

```python
import pandas as pd
import random
import requests
from PIL import Image
from io import BytesIO
meme_url = 'https://raw.githubusercontent.com/awchisholm/snippets/main/memes.csv'
memes = pd.read_csv(meme_url)
x = random.choice(memes['value'])
top = random.choice(lgbtq) + '  '+ random.choice(action)
bottom = random.choice(connective) + '  '+ random.choice(food)
URL = f'http://apimeme.com/meme?meme={x}&top={top}&bottom={bottom}'
meme = requests.get(URL)
img = Image.open(BytesIO(meme.content))
img.save('g.jpg')
img
```
